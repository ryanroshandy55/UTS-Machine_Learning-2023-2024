```python
# Importing libraries
import numpy as np
import pandas as pd

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

from google.colab import files
```

# A. Reading Data Memecah Attribute Menjadi 2 Variabel

```python
csv_url = 'https://raw.githubusercontent.com/isnanmulia/lecture-datasets/main/mobileprice_modified.csv'
datasets = pd.read_csv(csv_url)

datasets_attr = datasets[['battery_power', 'blue', 'clock_speed', 'dual_sim', 'fc', 'four_g',
       'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height',
       'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time', 'three_g',
       'touch_screen', 'wifi']]
datasets_attr.describe()
datasets_attr.info()
```

# B. Cleaning Process

```python
datasets_missing_values = datasets_attr
print(datasets_missing_values)
datasets_missing_values.isna().sum()
```

## Fill NaN values with median

```python
# Checking data and fill N/A values with median
print(datasets_missing_values.median())
datasets_fillna = datasets_missing_values.fillna(datasets_missing_values.median())
datasets_fillna.isna().sum()
datasets_fillna
```

# C. Standarized Datasets using MinMaxScaler and StandarizedScaler

```python
datasets_min_max = datasets_fillna
datasets_standarized = datasets_fillna
```

---


## MinMaxScaler


---

```python
# Scaling the attribute values
scaler = MinMaxScaler(feature_range = (0, 1))
scaled_dataset = scaler.fit_transform(datasets_min_max)
print(datasets_min_max.head(10))
scaled_dataset[:5,:]
print(datasets_min_max.describe())
```

---
## StandarizedScaler
---

```python
s_scaler = StandardScaler(with_mean = True)
print(s_scaler.fit(datasets_attr))
print(s_scaler.mean_)
print(s_scaler.transform(datasets_attr))
print(datasets_attr.describe())
```

```python
s_scaler = StandardScaler(with_mean = True)
print(s_scaler.fit(datasets_standarized))
print(s_scaler.mean_)
print(s_scaler.transform(datasets_standarized))
print(datasets_standarized.describe())
```
